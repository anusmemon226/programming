## Question 2:
How many missing values are there in each column?

```python
df.isna().sum()
```

## Question 3: 
Replace the missing laptop brands with the most frequent manufacturer brand.

```python
df['Manufacturer']=df['Manufacturer'].fillna(df['Manufacturer'].mode()[0])
df
```

## Question 4:
replace the missing values in the weight column with mean weight.

```python
df['Weight_kg']=df['Weight_kg'].fillna(df['Weight_kg'].mean())
df
```

## Question 5:
drop the further missing records if any.

```python
df.dropna(inplace=True)
```
